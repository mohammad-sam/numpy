```python
print(a.T)
```

```python
print(a[0:1]) 
print(a[0:1,2]) 
print(a[:,1])
```

```python
print(np.zeros((3, 4)))
print(np.ones((2, 3), dtype=np.int16))
print(np.random.rand(2,3))
print(np.eye(2))
```
